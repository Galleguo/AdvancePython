<H2> T- Test </H2>

¿Cuándo se usa?

- Cuando se quiere determinar si dos medias de dos grupos son iguales
- Cuando se quiere comparar dos medias de dos grupos distintos
- Se desconoce la varianza
- Las muestras deben ser menores o iguales a 30.
    Mayores a 30 se recomienda usar Z-Test
- La distribución de las muestras es normal

Existen 3 tipos distintos de T-Test para muestras: **Para muestras independientes, muestras emparejadas y una sola muestra.**

<H3>T-Test de muestras independientes:</H3>

1- Verificar el promedio de 2 grupos independientes no relacionados
2- Las muestras deben ser obtenidas de dos poblaciones distintas

Hipotesis nula: μa = μb. Con μ:media

Hipotesis alternativa: μa ≠ μb

```python
import random
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

#gauss(media,std)
#A continuación genero 2 muestras random con media y desviación distinta
a = [random.gauss(50,20) for x in range(30)] 
b = [random.gauss(55,15) for x in range(30)]

#print(len(a),a, len(b))

sns.set_style('darkgrid')
sns.kdeplot(a, fill = True, label = "a")
sns.kdeplot(b, fill = True, label = "b")
plt.legend()
plt.show()

#Independent T-Test
#Hipotesis nula: μa = μb

t_stat, p_value = stats.ttest_ind(a,b,equal_var=False) #equal_var=False no tengo en cuenta la varianza
if p_value > 0.05:
    print("Valor p:",p_value,". Como p_value es mayor a 0.05 se acepta la hipotesis nula de que ambas medias son casi iguales")
else:
    print("Valor p:",p_value,". Como p_value es menor a 0.05 no se acepta la hipotesis nula de que ambas medias son casi iguales")

#Vamos a confirmar la hipotesis:

print("Media grupo a:", np.mean(a), "Media grupo b:", np.mean(b))
```

<H3>T-Test de Muestras emparejadas:</H3>

1- Promedio de 2 muestras obtenidas de una misma población en distintos tiempos

Hipotesis nula: μa-μb = μd = 0

Hipotesis alternativa: μd ≠ 0

```python
a = [random.gauss(50,20) for x in range(30)] 
b = [random.gauss(60,25) for x in range(30)]

sns.set_style('darkgrid')
sns.kdeplot(a, fill = True, label = "a")
sns.kdeplot(b, fill = True, label = "b")
plt.legend()
plt.show()

#Paired T-Test
#Hipotesis nula: μa-μb = μd = 0

t_stat, p_value = stats.ttest_rel(a,b) #equal_var=False no tengo en cuenta la varianza
if p_value > 0.05:
    print("Valor p:",p_value,". Como p_value es mayor a 0.05 se acepta la hipotesis nula")
else:
    print("Valor p:",p_value,". Como p_value es menor a 0.05 no se acepta la hipotesis nula")

#Vamos a confirmar la hipotesis:

print("Diferencia de las medias:", np.mean(a)-np.mean(b))

```

<H3>T-Test de una muestra:</H3>

1- Promedio de un solo grupo es distinto del promedio conocido

Hipotesis nula: μa = X

Hipotesis alternativa: μa ≠ X

```python
a1 = [random.gauss(50,20) for x in range(30)] 

sns.set_style('darkgrid')
sns.kdeplot(a1, fill = True, label = "a1")
plt.legend()
plt.show()

#Paired T-Test
#Hipotesis nula: μa = X
#Vamos a forzar a que la hipotesis nula se cumpla

t_stat, p_value = stats.ttest_1samp(a1,np.mean(a1)) #equal_var=False no tengo en cuenta la varianza
if p_value > 0.05:
    print("Valor p:",p_value,". Como p_value es mayor a 0.05 se acepta la hipotesis nula")
else:
    print("Valor p:",p_value,". Como p_value es menor a 0.05 no se acepta la hipotesis nula")
```
