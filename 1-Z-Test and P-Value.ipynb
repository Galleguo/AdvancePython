<H2> Z-Test and P-Value </H2>
Z-Test es el equivalente al T-Test pero cuando el tamaño de las muestras es **MAYOR A 30**

¿Cuándo se usa?
- Varianza de la población es conocida ó
- Varianza de la población desconocida pero el tamaño de las muestras es mayor a 30


1 sample z-test<br>
2 sample z-test

<H3> One sample Z-test </H3>

Asumamos lo siguiente:<br>
Un doctor sabe que el IQ de las personas es de 100 con una desviación de 15. Este doctor creó una droga y quiere verificar si impacta en el IQ de las personas.

Va a tener por lo tanto dos grupos de muestras: Uno antes de probar la droga y otro después.


$ \mu = 100 $ <br>
$ \sigma = 15 $

```python
import random
random.seed(20)
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.stats.weightstats import ztest as ztest

#Hipotesis nula: mean = X (la droga no funcionó)

a = [random.gauss(100,15) for x in range(40)] #población conocida
sns.set_style('darkgrid')
sns.kdeplot(a, fill = True, label = "Población luego de consumir la droga")
plt.legend()
plt.show()

z_value, p_value = ztest (a, value=100)

if p_value > 0.05:
    print("Valor p:",p_value,". Como p_value es mayor a 0.05 se acepta la hipotesis nula de que la droga NO funcionó, mismo IQ")
else:
    print("Valor p:",p_value,". Como p_value es menor a 0.05 no se acepta la hipotesis nula de que la droga funcionó, distinto IQ")

#Hipotesis nula: mean = X (la droga no funcionó)
#Hipotesis alternativa: mean > X (la droga aumento el IQ)

a1 = [random.gauss(120,15) for x in range(40)] #población conocida
sns.set_style('darkgrid')
sns.kdeplot(a1, fill = True, label = "Población luego de consumir la droga")
plt.legend()
plt.show()

z_value1, p_value1 = ztest (a1, value=100, alternative = 'larger')

if p_value1 > 0.05:
    print("Valor p:",p_value1,". Como p_value es mayor a 0.05 se acepta la hipotesis nula de que la droga NO funcionó, mismo IQ")
else:
    print("Valor p:",p_value1,". Como p_value es menor a 0.05 no se acepta la hipotesis nula de que la droga funcionó, distinto IQ")

    
#Hipotesis nula: mean = X (la droga no funcionó)
#Hipotesis alternativa: mean < X (la droga disminuyó el IQ)

a2 = [random.gauss(80,15) for x in range(40)] #población conocida
sns.set_style('darkgrid')
sns.kdeplot(a2, fill = True, label = "Población luego de consumir la droga")
plt.legend()
plt.show()

z_value1, p_value2 = ztest (a2, value=100, alternative = 'smaller')

if p_value1 > 0.05:
    print("Valor p:",p_value1,". Como p_value es mayor a 0.05 se acepta la hipotesis nula de que la droga NO funcionó, mismo IQ")
else:
    print("Valor p:",p_value1,". Como p_value es menor a 0.05 no se acepta la hipotesis nula de que la droga funcionó, distinto IQ")
```

<H3> 2 Sample Z-Test </H3>

Vamos a comprobar si el IQ de dos poblaciones son similares

```python
#Hipotesis nula: mean a = mean b (ambas poblaciones tienen el mismo IQ)

a = [random.gauss(100,15) for x in range(40)]
b = [random.gauss(120,15) for x in range(40)]
sns.set_style('darkgrid')
sns.kdeplot(a, fill = True, label = "Población a")
sns.kdeplot(b, fill = True, label = "Población b")
plt.legend()
plt.show()

z_value, p_value = ztest (a,b, value=0)
#value = 0 es para indicar que la diferencia de ambas medias debe ser igual a 0 como definimos en la hipotesis nula

if p_value > 0.05:
    print("Valor p:",p_value,". Como p_value es mayor a 0.05 se acepta la hipotesis nula, mismo IQ")
else:
    print("Valor p:",p_value,". Como p_value es menor a 0.05 no se acepta la hipotesis nula, distinto IQ")
```
