<H2> <A HREF=https://scikit-learn.org/stable/> Scikit-Learn </A> 3.41 </H2>

Es una librería de Python de Aprendizaje Automático ó Machine Learning.

Es usada para automatizar completamente o parcialmente el análisis de datos y extracción de información.

Es utilizada para crear modelos de Machine Learning con capacidad de Clasificación, Regresión y Clustering.

Características:

- Herramientas para identificar, organizar y resolver problemas reales
- Descarga gratuita de datasets
- Librerías para aprender y realizar predicciones

Scikit-Learn ayuda a Data Scientis e Ingenieros de Machine Learning a resolver problemas utilizando la aproximación problema-solución.

Aproximación Problema-Solucion:
1- Seleccionar el modelo
2- Encontrar un objeto estimador
3- Entrenar el modelo
4- Realizar predicciones
5- Mejorar el modelo -> Mejorar precisión

Cuidados a tener en cuenta cuando se trabaja con este modelo:
- Crear objetos separados para las CARACTERÍSTICAS y las RESPUESTAS
- Asegurarse que las características y respuestas solo tienen valores numéricos.
- Verificar que las características y respuestas estén en el formato de NumPy array.
- Verificar que las características y respuestas tienen el mismo tamaño y forma en el array.
- Cuando se grafica. La abcisa x son las características y la respuesta la ordenada y.
