<H2>Series</H2>


```python
import pandas as pd

import numpy as np

 

l = [x for x in range(5)] #creo una lista de elementos de 0 a 4 incluido

lis2 = [0, 4, 8, 16, 3.5]

 

#Se puede transformar una lista en una serie

 

s_lis2 = pd.Series(lis2)

 

print("Se puede transformar una lista en una serie. La primera columna indica la etiqueta (distinto de indice) para acceder al valor \n",s_lis2, sep='') 
 

print(s_lis2[3])

 

#Se asigna de forma automatica la etiqueta de 0 a n (cantidad de elementos). Pero se puede definir el valor de la etiqueta

 

index = ["valor 1","valor 2", "valor 3", "valor 4", "valor 5"]

s_lis2_index = pd.Series(lis2,index)

print ("\nSe asigna de forma automatica la etiqueta de 0 a n (cantidad de elementos). Pero se puede definir el valor de la etiqueta \n", s_lis2_index, sep='')
print("Las etiquetas pueden ser repetidas. Cuando consultemos el valor de la etiqueta entonces va a dar los valores donde se encuentra esa etiqueta")

 

#También se puede crear una Serie a partir de un diccionario. Las etiquetas del diccionario se van a respetar en la Serie.

 

diccionario_prueba = {'a': 1, 'b':2, 'c':3, 'd':4, 'e':5}

serie_dic = pd.Series(diccionario_prueba)

print ("\nTambién se puede crear una Serie a partir de un diccionario. Las etiquetas del diccionario se van a respetar en la Serie.\n",serie_dic,sep='')

 

#Existe la posibilidad de filtrar y convertir en series unicamente las etiquetas que se indican

 

print("\nExiste la posibilidad de filtrar y convertir en series unicamente las etiquetas que se indican\n",pd.Series(diccionario_prueba, ['a','d']),sep='')

 

print("Si no existe la etiqueta asignada sea crea un valor no numérico (NaN)")
 

print("\nLa Lista: ", '\n', s_lis2_index, "\nEl valor en 0 accediendo por etiqueta:\n", s_lis2_index["valor 1"], sep='')

print("\nEl valor en 4 accediendo por ubicacion: ", s_lis2_index.iloc[3], "\nO también: ", s_lis2_index.iat[3], sep='')
```

<H3> Condicionales </H3>

Se puede realizar acciones sobre condiciones falsas con 'where'

```python
#Se puede asignar una variable o se puede realizar una operacion sobre el valor falso

lista_condicion = s_lis2_index.where(s_lis2_index%2==0, 'No es par') #si no se especifica completa con NaN

print ("Luego de la condicion:\n", lista_condicion, sep='')

print("\nRedondeo hacia arriba los que dan falso (combino con librerìa NumPy):\n", s_lis2_index.where(s_lis2_index%2==0, np.ceil(s_lis2_index)),sep='')

 
#Agregando inplace = True asigna automaticamente a la variable la condicion (sino se tieen que igualar a una variable)
s_lis2_index.where(s_lis2_index%2==0, inplace = True) 


print("\nAgregando inplace = True asigna automaticamente a la variable la condicion (sino se tieen que igualar a una variable)\n",s_lis2_index,sep='')


#Manejo de NaN 

print("\nCon .dropna() elimino los valores NaN\n", s_lis2_index.dropna(),sep='') #elimina los valores NaN

print("\nCon .fillna(), rellenos los valores NaN con lo que le paso como argumento\n",s_lis2_index.fillna("Numero impar"),sep='') #Rellena los valores NaN con el parametro que se pasa

```

<H3> DataFrame </H3>

Es un tipo de objeto de la librería Pandas

```python
import pandas as pd

df = pd.DataFrame()

print("Tipo de dato almacenado en df:",type(df),sep='\n')

df = pd.read_csv("PandasExample.csv")

print("\nEl archivo leído contiene la siguiente información:\n", df, sep='')

print("\nSe puede mostrar los primeros elementos con la función .head(), por default muestra los primeros 5:\n", df.head(),sep='')

print("\nSe puede mostrar los últimos elementos con la función .tail(), , por default muestra los últimos 5:\n", df.tail(),sep='')

print("\nSe puede mostrar los primeros n elementos con la función .head(n), en este caso se muestran los primeros 3:\n", df.head(3),sep='')

print("\nSe puede mostrar los últimos n elementos con la función .tail(n), en este caso se muestran los últimos 3:\n", df.tail(3),sep='')

#Acceder a cada conjunto de valores

print("\nDe igual forma que las Series, se puede acceder a cada conjunto de valores por la posición con .iloc(). Se muestra el valor en 0 y 4:\n",df.iloc[0],'\n\n',df.iloc[4],sep='')

#También se puede mostrar los valores de dataframe como una lista de arrays (definidos en NumPy), con df.values

#Si el archivo de lectura es muy largo se puede subdividir en varios dataframe con chunksize
#En chunksize = n se indica cada cuantas filas se agrupa cada dataframe

df_dividido = pd.read_csv("PandasExample.csv", chunksize = 3)
#cuando se usa chunksize se devuelve un puntero a la dirección de memoria del dataframe

for chunk in df_dividido:
    print(chunk)

#Se puede filtrar donde los valores sean verdaderos por etiqueta y condición según:
print("\nSe muestran las personas mayores de 21 años:\n", df[df['Age']>21])
```

<H2> Text Data (3.24) </H2>

La serie generada a partir de información brindada es del tipo 'object'. Eso quiere decir que, en el caso que se le pase información del tipo string, se deberá definir el tipo dtype = 'string' dentro del argumento.

Los textos incluidos dentro de las listas también pueden ser divididos con el método Serie.str.split(arg)

Los datos incluidos dentro de la Serie se pueden concatenar Series.str.cat(sep = 'arg')

```python
import pandas as pd

grades = pd.Series(['Romeo','Julieta','Mario','BadBunny','EminEm','eFe'], dtype='string')
#en la presentación escriben dtype dentro de la lista que se pasa como parámetro

print(grades.str.upper(),'\n', grades.str.lower(), sep='')

lista_separada = grades.str.split('o') #crea un dtype = object
print(lista_separada)

lista_unida = grades.str.cat(sep = ' ') #concatenate method
print(lista_unida, type(lista_unida))
```
