<H1> Data Visualization </A> </H1>

Data visualization (visualización de datos): es la técnica que permite presentar los datos en un formato pictorial o gráfico.

Ventajas:

- Simplifica la información cuantitativa compleja
- Analiza y explora Big Data de forma sencilla
- Identifica áreas para mejorar
- Identifica las relaciones entre los datos y las variables
- Permite explorar nuevos y ocultos patrones

Librería para visualizar data:

- matplotlib
- pygal

- networkx
- vispy
- bokeh
- seaborn

<H3> Crear una gráfica </H3>

```python
#Importar la librería

import numpy as np #se importa numpy para generar datos aleatorios
import matplotlib.pyplot as plt
from matplotlib import style #se usa para definir el stilo de la grilla (gride)
#%matplotlib inline (no es necesario en jupyter notebbok. Haciendo esto los gráficos salen en output y no en ventanas emergentes nuevas)

#Se define el dataset requerido
random_numbers = np.random.rand(10)
random_numbers_2 = np.random.rand(10)

#Setear las características del gráfico
style.use('ggplot') #defino el estilo de la grilla
plt.plot(random_numbers,'g', label ='line one',linewidth=2) #'g' ->color, linewidth -> expesor
plt.xlabel('Rango')
plt.ylabel('Números')
plt.title('Primer Plot')

#Mostrar el plot creado
plt.legend()
plt.show()
```

Resumiendo:
1-Cargar base de datos
2-Procesar info
3-Definir características del gráfico (label, legend, título, tamaño, etc)
4-Guarda (caso de que se requiera)
5-Mostrar

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('SalesExample.csv',parse_dates=['StartDate'],index_col=['StartDate'])
#parse_dates es para indicarle cual columna es de fechas.
df.sort_index(inplace=True)
#ordena las fechas de menor a mayor

print(df.columns)

plt.figure(
    figsize=(15,10),
    dpi=200)
plt.title('Gráfico de línea')
plt.xlabel('Fecha')
plt.ylabel('QuotaAmount')
plt.plot(df['QuotaAmount'])
plt.savefig('Nombreejemplo.jpg')
plt.show()

plt.boxplot(df['QuotaAmount'])
plt.show()

plt.hist(df['QuotaAmount'])
plt.show()

plt.xlabel('Quota Amount')
plt.ylabel('Name')
plt.title('Quota Amount by User')
plt.scatter(x=df['QuotaAmount'],y=df['OwnerName'])
plt.show()

dic = {'a':10, 'b':20, 'c':13}
plt.bar(x=dic.keys(), height = dic.values())
plt.show()

plt.pie(x=dic.values(), labels=dic.keys())
plt.show()
```

<H2> <A HREF=https://seaborn.pydata.org/> Seaborn </A> (3.34) </H2>

Seaborn es una librería que se utiliza junto con matplotlib para mostrar gráficos estadísticos de forma atractiva.

Se suele importar con el alias 'sns'. 
    
    Import seaborn as sns
   
Se pueden realizar todas las mismas gráficas que matplotlib, pero más atractivas

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()#seteo el tema por defecto
#fmri es un dataframe built-in de seaborn que tiene 1064 filas y 5 columnas
#las columnas son subject,timepoint, event, region y signal

fmri = sns.load_dataset("fmri") #se carga una base de datos incluida dentro de seaborne

print(fmri.head())

sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", col="region",
    hue="event", style="event"
) #es una funciòn que ya calcula medias y desvìos

plt.show()
```
